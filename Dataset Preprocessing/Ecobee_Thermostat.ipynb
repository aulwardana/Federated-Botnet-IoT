```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/benign_traffic.csv')
df2 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/gafgyt/combo.csv')
df3 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/gafgyt/junk.csv')
df4 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/gafgyt/scan.csv')
df5 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/gafgyt/tcp.csv')
df6 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/gafgyt/udp.csv')
df7 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/mirai/ack.csv')
df8 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/mirai/scan.csv')
df9 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/mirai/syn.csv')
df10 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/mirai/udp.csv')
df11 = pd.read_csv('/home/administrator/dataset/NBaIoT/Ecobee_Thermostat/mirai/udpplain.csv')
```

```python
df1['Label'] = 0
df1.info(verbose=True, show_counts=True)
```

```python
df_atc = pd.concat([df2, df3, df4, df5, df6, df7, df8, df9, df10, df11])
df_atc['Label'] = 1
df_atc.info(verbose=True, show_counts=True)
```

```python
df = pd.concat([df1, df_atc])

df.info(verbose=True, show_counts=True)
```

```python
df['Label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.3, random_state=100)
```

```python
pd.DataFrame(train).to_csv("./Ecobee_Thermostat_train.csv", index=False)
```

```python
pd.DataFrame(test).to_csv("./Ecobee_Thermostat_test.csv", index=False)
```
